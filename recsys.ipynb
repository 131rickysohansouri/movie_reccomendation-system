```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head()
```

```python
credits.head()
```

```python
#Joining the both datasets

data=movies.merge(credits,on="title")
data.head()
```

```python
data.info()
```

```python
#generes
#id
#keywords
#title
#overview
#cast
#crew

data=data[["id","title","overview","genres","keywords","cast","crew"]]
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.duplicated().sum()
```

```python
data.iloc[0].genres
```

```python
import ast

def transform(obj):
    
    temp=[]
    for i in ast.literal_eval(obj):
        temp.append(i['name'])
        
    return temp

data['genres'] = data["genres"].apply(transform)
data['keywords'] = data["keywords"].apply(transform)
```

```python
data.head()
```

```python
import ast

def transform_cast(obj):
    counter = 0
    temp=[]
    for i in ast.literal_eval(obj):
        if counter != 3:
            temp.append(i['name'])
            counter = counter+1
            
        else:
            break
    return temp

data['cast'] = data["cast"].apply(transform_cast)
```

```python
import json

def transform_cast(obj):
    temp=[]
    for i in json.loads(obj):
        if i["job"] == "Director":
            temp.append(i['name'])
            break
    return temp

data["crew"]=data["crew"].apply(transform_cast)
```

```python
data.head()
```

```python
data["genres"]=data["genres"].apply(lambda x:[i.replace(" ","") for i in x])
data["cast"]=data["cast"].apply(lambda x:[i.replace(" ","") for i in x])
data["crew"]=data["crew"].apply(lambda x:[i.replace(" ","") for i in x])
data["kewwords"]=data["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
data['overview']=data['overview'].apply(lambda x:x.split())
```

```python
data.head()
```

```python
data['tags']=data['overview']+data['genres']+data["keywords"]+data['cast']+data['crew']

```

```python
fdata=data[['id','title','tags']]
```

```python
fdata['tags']=fdata['tags'].apply(lambda x:" ".join(x))
fdata['tags']=fdata['tags'].apply(lambda x: x.lower())
```

```python
fdata.head()
```

```python
#Stemming
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stemming(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
fdata['tags']=fdata['tags'].apply(stemming)
```

```python
fdata.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=2500,stop_words='english')
```

```python
vectors=cv.fit_transform(fdata['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim=cosine_similarity(vectors)
```

```python
sim.shape
```

```python
def reccomend(movie):
    index=fdata[fdata["title"] == movie].index[0]
    distances = sim[index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(fdata.iloc[i[0]].title)
        
```

```python
reccomend("John Carter")
```
